Transformation des valeurs catégorique en numérique

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
data.head(3)
```

Avec label encoder

```python
data_lb=data[['survived','pclass','age','sibsp','parch','fare']].copy()
```

```python
le =LabelEncoder()
data_lb['Lb_sex']=le.fit_transform(data['sex'])
data_lb['Lb_embarked']=le.fit_transform(data['embarked'])
data_lb['Lb_class']=le.fit_transform(data['class'])
data_lb['Lb_who']=le.fit_transform(data['who'])
data_lb['Lb_adult_male']=le.fit_transform(data['adult_male'])
data_lb['Lb_deck']=le.fit_transform(data['deck'])
data_lb['Lb_embark_town']=le.fit_transform(data['embark_town'])
data_lb['Lb_alive']=le.fit_transform(data['alive'])
data_lb['Lb_alone']=le.fit_transform(data['alone'])
data_lb.head(2)
```

Avec OneHotEncoder

```python
data_ec=data[['survived','pclass','age','sibsp','parch','fare']].copy()
```

```python
transformer = make_column_transformer((OneHotEncoder(), ["sex","embarked","class","who","adult_male","deck","embark_town","alive","alone"]),remainder='passthrough')
transformed = transformer.fit_transform(data)
transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names())
```

```python
data
```

```python
transformed_df
```

Prediction: avec entrées avec valeurs nulles enlevé

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
```

```python
X_lb=data_lb[['Lb_sex', 'Lb_embarked', 'Lb_who', 'Lb_embark_town','pclass','age','sibsp','parch','fare','Lb_class','Lb_adult_male','Lb_deck','Lb_alone']]
y_lb=data_lb[['survived']]
```

```python
X_ohc=transformed_df.drop(["survived"],axis=1)
y_ohc=transformed_df[['survived']]
```

Avec LabelEncode

```python
X_train,X_test,y_train,y_test= train_test_split(X_lb,y_lb,test_size=0.3,random_state=12)
```

```python
model_xgb_lb= XGBRegressor()
```

```python
model_xgb_lb.fit(X_train,y_train)
```

```python
ypred_lb=model_xgb_lb.predict(X_test)
```

```python
dftemp_lb=y_test
dftemp_lb['ypred']=ypred_lb
```

```python
dftemp_lb.head(10)
```

Avec OneHotEncoder

```python
X_train,X_test,y_train,y_test= train_test_split(X_ohc,y_ohc,test_size=0.3,random_state=12)
model_xgb_ohc= XGBRegressor()
model_xgb_ohc.fit(X_train,y_train)
ypred_ohc=model_xgb_ohc.predict(X_test)
dftemp_ohc=y_test
dftemp_ohc['ypred']=ypred_ohc
dftemp_ohc.head(10)
```

```python
from sklearn.metrics import r2_score
print(r2_score(dftemp_lb["survived"],dftemp_lb["ypred"]))
print(r2_score(dftemp_ohc["survived"],dftemp_ohc["ypred"]))
```

Test sur tout le data

```python
ypred_lb_all=model_xgb_lb.predict(X_lb)
ypred_ohc_all=model_xgb_ohc.predict(X_ohc)
```

```python
print(r2_score(data_lb['survived'],ypred_lb_all))
print(r2_score(transformed_df['survived'],ypred_ohc_all))
```

```python
print("Resultat score r2 avec lalbel encoder",round(r2_score(data_lb['survived'],ypred_lb_all) * 100, 2))
print("Resultat score r2 avec OneHot encoder",round(r2_score(transformed_df['survived'],ypred_ohc_all) * 100, 2))
```
